```python
avg = 20
coefficient = 0.3
print("Coefficient:", coefficient)
errors = [2, -3, 5, -2, 1]
```

```python
predicted = [avg]

for i in range(1, len(errors)):
    new_predicted = avg + coefficient * errors[i-1]
    predicted.append(new_predicted)

print("Predicted:", predicted)
```

```python
# Tabulate results
rows = []
for step, err, pred in zip(range(len(errors)), errors, predicted):
    rows.append([step, err, pred])

print(tabulate(rows, headers=["Step","Error","Prediction"], tablefmt="grid"))
```

```python
# Plotting
t_values = list(range(1, len(errors)+1))
plt.plot(t_values, [avg]*len(errors), label="Average (U)", linestyle="--", color="red")
plt.plot(t_values, predicted, label="Predicted (J values)", marker="o", color="blue")

plt.xlabel("Time Period (t)")
plt.ylabel("Values")
plt.title("Average (U) vs Predictions (J)")
plt.legend()
plt.grid(True)
plt.show()
```
